```python
count  = 0
for i in df['gender'].unique():
    count = count + 1
    print(count,'. ',i)
```

```python
gender = df['gender'].value_counts()
plt.figure(figsize=(7, 6))
ax = gender.plot(kind='bar', rot=0, color="c")
ax.set_title("Male vs Female Students", y = 1)
ax.set_xlabel('Gender')
ax.set_ylabel('Number of People')
ax.set_xticklabels(('male', 'female'))

for rect in ax.patches:
    y_value = rect.get_height()
    x_value = rect.get_x() + rect.get_width() / 2
    space = 1
    label = format(y_value)
    ax.annotate(label, (x_value, y_value), xytext=(0, space), textcoords="offset points", ha='center', va='bottom')    
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 10))
fig.suptitle('Comparing Male & Female scores for all three subjects:', fontsize=20) 
# Math score
sns.boxplot(x='gender', y='math score', data=df, ax=ax1,palette = 'mako')
ax1.title.set_text('Math score vs Gender')
# Reading Score 
sns.boxplot(x='gender', y='reading score', data=df, ax=ax2,palette = 'mako')
ax2.title.set_text('Reading score vs Gender')
# Writing Score
sns.boxplot(x='gender', y='writing score', data=df, ax=ax3,palette = 'mako')
ax3.title.set_text('Writing score vs Gender')
```


In Math, on average, Male students have scored more than Female students. And one student that scored zero in Maths is a Female.

In Reading, female students have done better than male students. Also, the student that scored the lowest marks is a female.

Even for Writing, females have performed better than males and again the student that scored the lowest in this subject is also a female.


```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 10))
fig.suptitle('Comparing Race/Ethnicity vs scores for all three subjects:', fontsize=20)
# Math
sns.boxplot(x='race/ethnicity', y='math score', data=df, ax=ax1,palette = 'mako')
ax1.title.set_text('Math score vs race/ethnicity')
# Reading
sns.boxplot(x='race/ethnicity', y='reading score', data=df, ax=ax2,palette = 'mako')
ax2.title.set_text('Reading score vs race/ethnicity')
# Writing
sns.boxplot(x='race/ethnicity', y='writing score', data=df, ax=ax3,palette = 'mako')
ax3.title.set_text('Writing score vs race/ethnicity')
```


In Math, group E has performed the best and group A the worst. The student who scored zero belongs to group C.

In Reading, group E has performed the best and group A the worst. The student who scored the lowest marks belongs to group C.
In Writing, groups C, D & E have done well. Group A performed the worst and the lowest marks are from group C.
Students in group B performed average in all three subjects.

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
fig.suptitle('Comparing Parents education vs scores for all three subjects:', fontsize=20)
# Math
sns.barplot(x='parental level of education', y='math score', data=df, ax=ax1,palette = 'mako')
ax1.title.set_text('Math score vs Parents education')
var1 = ax1.set_xticklabels(ax1.get_xticklabels(), rotation=30)
# Reading
sns.barplot(x='parental level of education', y='reading score', data=df, ax=ax2,palette = 'mako')
ax2.title.set_text('Reading score vs Parents education')
var2 = ax2.set_xticklabels(ax2.get_xticklabels(), rotation=30)
# Writing
sns.barplot(x='parental level of education', y='writing score', data=df, ax=ax3,palette = 'mako')
ax3.title.set_text('Writing score vs Parents education')
var3 = ax3.set_xticklabels(ax3.get_xticklabels(), rotation=30)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
fig.suptitle('Comparing Test Preparation Course vs scores for all three subjects:', fontsize=20)
# Math
sns.barplot(x='test preparation course', y='math score', data=df, hue='gender',ax=ax1,palette = 'mako')
ax1.title.set_text('Math score vs Parents education')
var1 = ax1.set_xticklabels(ax1.get_xticklabels(), rotation=30)
# Reading
sns.barplot(x='test preparation course', y='reading score', data=df,  hue='gender',ax=ax2,palette = 'mako')
ax2.title.set_text('Reading score vs Parents education')
var2 = ax2.set_xticklabels(ax2.get_xticklabels(), rotation=30)
# Writing
sns.barplot(x='test preparation course', y='writing score', data=df,  hue='gender',ax=ax3,palette = 'mako')
ax3.title.set_text('Writing score vs Parents education')
var3 = ax3.set_xticklabels(ax3.get_xticklabels(), rotation=30)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
fig.suptitle('Comparing Parents education vs scores for all three subjects:', fontsize=20)
# Math
sns.barplot(x='parental level of education', y='math score', data=df, ax=ax1,palette = 'mako')
ax1.title.set_text('Math score vs Parents education')
var1 = ax1.set_xticklabels(ax1.get_xticklabels(), rotation=30)
# Reading
sns.barplot(x='parental level of education', y='reading score', data=df, ax=ax2,palette = 'mako')
ax2.title.set_text('Reading score vs Parents education')
var2 = ax2.set_xticklabels(ax2.get_xticklabels(), rotation=30)
# Writing
sns.barplot(x='parental level of education', y='writing score', data=df, ax=ax3,palette = 'mako')
ax3.title.set_text('Writing score vs Parents education')
var3 = ax3.set_xticklabels(ax3.get_xticklabels(), rotation=30)
```
